```python
%matplotlib notebook 

import numpy
import matplotlib.pyplot as plot
```

```python
import matplotlib
matplotlib.rc('text', usetex=True)
```

```python
def zo_loss(s):
    return numpy.maximum(0., numpy.sign(-s))
```

```python
def log_loss(s):
    return numpy.log(1 + numpy.exp(-s)) / numpy.log(2)
```

```python
def hinge_loss(s):
    return numpy.maximum(0, 1 - s)
```

```python
xn = numpy.linspace(-5, 0, 100, endpoint=False)
xp = numpy.linspace(0, 5, 100)

plot.figure()

plot.hold('on')

plt0, = plot.plot(xn, zo_loss(xn), 'k-')
plt0, = plot.plot(xp, zo_loss(xp), 'k-')
plt1, = plot.plot(xn, log_loss(xn), 'r--')
plt1, = plot.plot(xp, log_loss(xp), 'r--')
plt2, = plot.plot(xn, hinge_loss(xn), 'b-.')
plt2, = plot.plot(xp, hinge_loss(xp), 'b-.')


plot.hold('off')


plot.legend([plt0, plt1, plt2], [
        '0-1 Loss', 'Log Loss', 'Hinge Loss'],
           loc='best')
plot.grid('on')
plot.xlim([-5, 5])
plot.ylim([-0.1, 7])

plot.xlabel('score')

plot.show()
```
