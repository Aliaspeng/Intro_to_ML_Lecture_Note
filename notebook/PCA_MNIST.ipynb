```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.decomposition import PCA
from sklearn.datasets import fetch_mldata
```

```python
mnist = fetch_mldata('MNIST original')
```

```python
X = mnist.data
y = mnist.target
print ('X Shape: ', X.shape)
```

```python
plt.imshow(X[100].reshape((28, 28)))
```

```python
pca = PCA(n_components=2)

X_reduced = pca.fit_transform(X)
```

```python
X_reduced.shape
```

```python
plt.figure(figsize=(17, 15))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c = y)
plt.colorbar()
```

```python
# Important thing - How much variance is explained by each dimension

print ('Ratio of variance explained: ', pca.explained_variance_ratio_)
```

## Centering

```python
X_centered = X - X.mean(axis = 0)
```

```python
# Verifying the centering
print np.sum(X.mean(axis=0))
print np.sum(X_centered.mean(axis=0))
```

```python
X_centered_reduced = pca.fit_transform(X_centered)
```

```python
plt.figure(figsize=(15, 15))
plt.scatter(X_centered_reduced[:, 0], X_centered_reduced[:, 1], c = y)
```

```python
print ('Ratio of variance explained: ', pca.explained_variance_ratio_)
```
